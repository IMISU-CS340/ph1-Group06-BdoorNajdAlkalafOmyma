## 1. المقدمة

### 1.1. الأهداف

هذه المفكرة موجهه لطلاب علوم الحاسب وأي شخص مهتم بخوارزميات الذكاء الاصطناعي. إيجاد حل للمشكلات باستخدام خوارزميات البحث المستنير وخوارزميات البحث المحلي.

### 1.2. المفاهيم الأساسية

#### 1.2.1. وصف المشكلة

##### الحالة الأولية
المدينة رقم 0 ؛ وهي المدينة الحالية للبائع المتجول.

##### الاجراءات
السفر من مدينة إلى أخرى من خلال هذه الإجراءات:
- اذهب يمينا
- أذهب يسارا
- اذهب للأعلى
- انزل
- اذهب لأعلى اليمين
- اذهب لأعلى اليسار
- اذهب لأسفل اليمين
- اذهب لأسفل اليسار

##### نموذج الانتقال
بالنظر إلى الإجراءات والحالات ، سيؤدي ذلك إلى إرجاع الحالة الناتجة ،
على سبيل المثال ، إذا سافر البائع من المدينة 0 إلى المدينة 1 ، فسيؤثر ذلك في تكلفة المسار وسيكون لديه المزيد من الخيارات للمدن المختلفة للسفر.

##### فضاء الحاله
عدد من المدن المختلفة ؛ بناءً على عدد المدن التي يدخلها المستخدم.

##### اختبار الهدف
قيام البائع المتجول بزيارة كل مدينة مرة واحدة فقط, برحلته التي تبدأ وتنتهي عند المدينة 0

#### 1.2.2. البحث المستنير
البحث المستنير يستخدم المعرفة المحددة للمشكلة عوضاً عن وصفها. وطرق البحث المستنير يجب ان يكون لها وصول إلى معادلة التوقع، حيث أنها تتوقع تكلفة الحل.  

##### معادلة التوقع

معادلة التوقع وتسمى اختصاراً التوقع، تستخدم لحساب قيمة التوقع، حيث أن هذه القيمة تعطي الخوارزمية التكلفة المتوقعة للطريق من النقطة الحالية وحتى نقطة الهدف. بمعنى آخر يمكن أن القول أنها تخبر الخوارزمية بأفضل طريق للوصول بأقصى سرعة ممكنة. 

إن جودة التوقع لخوارزمية البحث يعتمد على جودة التوقع نفسها. يمكن أحياناً بناء توقع جيد عبر تبسيط تعريف ووصف المشكلة أو عبر حفظ تكلفة حلول المشاكل البسيطة التي تم حسابها مسبقاً حيث يتم حفظها في قاعدة بيانات أو عبر التعلم من التجارب السابقة التي في نفس تصنيف المشكلة.

ان معادلات التوقع هي أكثر الطرق شهرة في اضافة المعرفة للمشكلة وونقلها وقيادتها.   

##### A*

اي نجمة هي أكثر خوارزميات البحث نجاحاً في ايجاد أقصر طريق بين مجموعة نقاط. هي ايضاً خوارزمية بحث مستنير حيث أنها تستخدم المعلومات عن تكلفة الطريق والتوقع لإيجاد الحل. 

بحث الاي نجمة يستكشف النقاط ذات الأقل قيمة لـf(n) حيثf(n)=g(n)+h(n).أيضاً هو بحث يحقق المثالية والاكتمال وهي خصائص قيّمة لخوارزميات البحث. المثالية تعني الضمان لايجاد أفضل حل ممكن، بينما الاكتمال يعني انه في حال وجود حل للمشكلة فهو يضمن إيجاده.  

لفهم كيفية عمل A* يجب علينا أولاً التعرف على بعض المفاهيم:

* النقطة:هي كل نقطة توقف محتمله ولها تعريف خاص.
* الانتقال: هو التحرك ما بين نقطتين.
* نقطة البداية: هنا حيث يبدأ البحث.
* نقطة الهدف: هي النقطة التي ينتهي البحث عند الوصول إليها.
* فضاء البحث: هو مجموع كل النقاط.
* التكلفة: هي قيمة رقمية للطريق المؤدي من نقطة إلى اخرى.
* g(n): هو القيمة الفعلية للطريق من نقطة البداية إلى اي نقطة اخرى.
* h(n):هو التوقع للتكلفة المتطلبة للطريق من النقطة n إلى نقطة الهدف.
* f(n): هو أقل تكلفة ما بين النقاط المجاوره.

#### 1.2.3. البحث المحلي

##### الاكتمال مقابل تكوينات المشكلة المتدرجة 

ما معنى ان نقول بان خوارزمية البحث متكاملة او انها متدرجة؟

إذا كانت خوارزمية البحث متكاملة فهذا يعني باننا نستطيع إيجاد حل واحد على الأقل عندها فان الخوارزمية ستضمن لنا إيجاد حل في وقت محدد. اما إذا كانت خوارزمية البحث متدرجة فهذا يعني بانها تعيد استخدام معلومات من عمليات بحث سابقة لإيجاد حلول لسلسلة من عمليات البحث المماثلة من الممكن جعله اسرع مما هو عليه عن طريق حل كل مشكلة بحث من البداية.  

##### خوارزمية Simulated Annealing:

اشتق هذ الاسم من طريقة التحسين التي تحاكي التبريد البطيء للمعادن والذي يتميز بانخفاض تدريجي في الحركات الذرية التي تقلل من كثافة العيوب الشبكية حتى يتم الوصول إلى حالة الطاقة الأقل.

الخوارزمية هي تقنية احتمالية لتقريب المثالية العامة لوظيفة معينة على وجه التحديد، إنها عملية تجريبية لتقريب المثالية العامة في مساحة بحث كبيرة لمشكلة التحسين، يتم استخدامها غالبًا عندما تكون مساحة البحث منفصلة وتستخدم هذه الخوارزمية لمشكلات البحث التي يكون فيها إيجاد حل عام قريب للحل المثالي اهم من إيجاد حل مثالي و دقيق لكنه محلي.

## 2. مثال: مشكلة البائع المتجول - Traveling Salesman Problem (TSP) 

### 2.1. وصف المشكلة
هي مشكلة لبائع يريد زيارة العديد من الزبائن القانطين في مدن مختلفة ويحاول العثور على اقصر طريق لتحقيق هذه الزيارات.

المشكلة تحاول العثور على دورة هاملتونين ذات الأقل تكلفة للطريق في الرسمة. اما مشكلة دورة هاملتونين هي إيجاد ما اذا كان هناك دورة تمر على جميع المدن مرة واحدة فقط. مشكلة البائع المتجول هي مشكلة NP صعبة مشهورة.

### 2.2. صياغة المشكلة
يكون معطى مجموعة من المدن، حدود موزونة وتمثل المسافة بين كل زوج من المدن، وأيضاً رسم متجة. ويتم لعمل حتى إيجاد أقصر طريق دائري يمر على جميع المدن في الرسم مرة واحدة فقط.

## 3. خوارزمية A*

### 3.1. وصف الخوارزمية
خوارزمية البحث A* هي خوارزمية اكتشاف المسار تستخدم في الذكاء الاصطناعي وتستخدم لحل مشاكل البحث عن أقصر مسار وتستخدم أيضًا للمساعدة في حل المشكلات المتعلقة بالألعاب للعثور على أقصر مسار بطريقة فعالة.
كيف تعمل هذه الخوارزمية؟
أولاً ، باعتبار أن لدينا رسم بياني بعقد متعددة (انظر الشكل 1). تحتوي كل عقدة على معادلة التوقع  (معادلة التوقع هي وظيفة تقوم بتصنيف البدائل في خوارزميات البحث في كل خطوة متفرعة بناءً على المعلومات المتاحة لتحديد الفرع الذي يجب اتباعه) وهناك قيمة المسافة بين كل عقدتين لديهم مسار (تسمى " تكلفة المسار ").
 هدفنا هو الوصول إلى العقدة النهائية (الهدف) بدءًا من العقدة الأولى في أقرب وقت ممكن.


A-star.png

تستخدم الخوارزمية هذه المعادلة للعثور على المسار الأقصر:

f(n)= g(n) + h(n) 

حيث أن:

n: العقدة السابقة على المسار.

g (n): تكلفة المسار من عقدة البداية إلى n.

h(n):معادلة التوقع التي تقدر تكلفة أقل مسار من العقدة n  إلى عقدة الهدف (النهاية).

لذلك ، تزور الخوارزمية العقدة الأولى وتطبق المعادلة السابقة ؛ فتحسب f (n) لكل عقدة متفرعة منه  ، ثم يأخذ أصغر f (n) منهم ويختارها لتكون عقدة البداية الجديدة.
ثم تقوم بتطبيق نفس الخطوات على عقدة البداية الجديدة (مع الملاحظة عند حساب قيمة g (n) , نضيف جميع تكاليف المسارات بدءًا من العقدة الأولى) ثم تختارعقدة بداية جديدة وتكررنفس الخطوات السابقة حتى تصل إلى العقدة النهائية.

### 3.2. تطبيق الخوارزمية

```python
#This is an implementation and applying of A* algorithm on TSP.
# Modules
import random
import time

class TSP(object):
    
    #Generates distance between 2 points
    def getDistance(self,P1,P2): 
        self.P1 = P1
        self.P2 = P2
        distance = ((self.P1[0]-self.P2[0])**2 + (self.P1[1]-self.P2[1])**2)**(1/2)
        return distance

    def generateCoordinates(self):
        x = random.randint(0,101)
        y = random.randint(0,101)
        return [x,y]

    #Calculates the total distance in a state, eg [0,1,2,3]
    def calculateDist(self,s,n): 
        self.calD = s
        self.nu = n
        dist = 0
        total = 0  
        for i in range(self.nu):
            xi = self.calD[i]
            yj = self.calD[i+1]
            dist = self.getDistance(self.coord[xi],self.coord[yj])
            total+=dist    
        return total
    
    #Generates the total  heuristic plus path cost and sends it out
    def hue(self,chosenList,cityList): 
        toCheck = cityList[:]
        SPL = chosenList[:]
        dl = 999999999
        fCost =[]
        l=[]
            
        for i in chosenList:
                toCheck.remove(i)
            
        for e in toCheck:
            l.clear()
            l.append(e)
            rest = cityList[:]
            totalDist = 0
            
            while len(rest) > 0:
                dl=99999
                for i in l:
                    if i in rest:
                        rest.remove(i)
                
                for n in l:
                    for m in rest:
                        d = self.getDistance(self.coord[n],self.coord[m])
                        if d<dl:
                            dl = d
                            c = m
                
                if c not in l:
                    l.append(c)
                    
                totalDist += d
            
            g = self.getDistance(self.coord[e],self.coord[SPL[-1]])
            k = self.getDistance(self.coord['0'],self.coord[e])
            f = g+totalDist+k
            
            fCost.append((f,e))
            
        fCost.sort()

        return (fCost[0][1])

    def solver(self): 
        
        #Take number of cities from the user
        print("أدخل عدد المدن المراد انشاءها:")
        number = input()  
        self.coord ={}
        cityList = []
        chosenList = ['0']
        self.number = int(number)
        for i in range(self.number):
            a = str(i)
            l = self.generateCoordinates()
            self.coord[a] = l
            cityList.append(a)
        
        currentState = cityList[:]
                
        for i in range(len(cityList)-1):        
            x = self.hue(chosenList,cityList)
            chosenList.append(x)
        
    
        final = chosenList + ['0']
        il = cityList + ['0']
        
        fd = self.calculateDist(final,len(final)-1)
        id = self.calculateDist(il,len(il)-1)
        
        print("\nالحالة الابتدائية تبدو هكذا:\n",il,"\n")
        print("المسافة الابتدائية كانت تساوي %.2f كم\n" %id)
        print("الحالة النهائية تبدو هكذا:\n",final,"\n" )
        print("المسافة الامثل تساوي %.2f كم \n" %fd)
  
         
def main():
    start = time.time()
    tsp = TSP() 
    tsp.solver()
    end = time.time()
    print("مقياس الوقت: %.2f ثانية" % (end-start))

if __name__ == "__main__":
    main()          
            
```

## 4. خوارزمية Simulated Annealing

### 4.1. وصف الخوارزمية
محاكاة عملية التلدين :

مستوحاة من الفيزياء والقوانين الفيزيائية وتُستخدم خوارزمية محاكاة عملية التلدين في حل العديد من مسائل (مشاكل التحسين) حيث
تعتمد الخوارزمية على طريقة عشوائية حسابية بهدف ايجاد الحلول القصوى – العليا والدنيا – للعديد من المسائل. في الواقع، تكمن قوة
خوارزمية محاكاة التلدين في قدرتها على تجنب السقوط بالنهايات العظمى المحلية وتكا د تكون الحلول التي تقدمها الخوارزمية افضل
من غيرها من الخوارزميات، ولكنها ليست الأفضل على الاطلاق. يمكن تصور ذلك عن طريق تخيل رسم ثنائي البعد كما هو موضح
بالشكل ادناه, حيث تمثل كل نقطة من الاحداثيات الواقعة على محور س حلا من الحلول ، وتمثل كل نقطة احداثيات على محور ص

كيف تعمل الخوارزمية :
1. في البداية, توليد حل عشوائي.
2. حساب الكلفة باستخدام احد التوابع التي قمت بتعريفها .

6. اعادة الخطوات من الثالثة حتى الخامسة حتى الوصول الى حل مقبول, او تصل الخوارزمية الى الحد الاعلى المسموح من التكرارات .

### 4.2. تطبيق الخوارزمية

```python
#This is an implementation and applying of SA algorithm on TSP.
# Modules
import time
import math
import numpy as np
import matplotlib.pyplot as plt

global input_data

# calulate length between two points
def length (n1,n2):
    return math.sqrt((n1[0] - n2[0])**2 + (n1[1] - n2[1])**2)

# calculate total length to traverse all points
def total_length(arr,n):
    l = length(arr[0],arr[n-1])
    for i in range(n-1):
        l+= length(arr[i],arr[i+1])
    return l

# two_opt optimization for simulated annealing, using a random probabilty function to do selection
def two_opt_optimization(sol_arr,t,n):

    # picking two pair of consecutive integers, making sure they are not same
    ai = np.random.randint(0,n-1)
    bi = (ai+1)%n 
    ci = np.random.randint(0,n-1)
    di = (ci+1)%n

    if ai != ci and bi != ci:
        a = sol_arr[ai]
        b = sol_arr[bi]
        c = sol_arr[ci]
        d = sol_arr[di]

        # old lengths
        ab = length(a,b)
        cd = length(c,d)
        # new lengths, if accepted by our probability function
        ac = length(a,c)
        bd = length(b,d)

        diff = ( ab + cd ) - ( ac + bd )

        p = 0
        # for negative diff-> we'll use boltzman probabilty distribution equation-> P(E)=exp(-E/kT)
        # k is considered to be 1
        if diff < 0:
            p = math.exp( diff/t )

        # we'll sometimes skip the good solution
        elif diff > 0.05 :
            p = 1

        if(np.random.random() < p ):
            new_arr = list(range(0,n))
            new_arr[0] = a
            i = 1

            while bi!= ci:
                new_arr[i] = c
                i = i+1
                ci = (ci-1)%n

            new_arr[i] = sol_arr[bi]
            i = i+1

            while ai!= di:
                new_arr[i] = sol_arr[di]
                i = i+1
                di = (di+1)%n

            return new_arr

    return sol_arr

# Simmulated Annealing algorithm
def sa_algorithm (input_data):

    #length of input_data
    n = len(input_data)

    #creating a base solution
    sol_arr = input_data

    #plot initial solution
    plt.axis([-100,1100,-100,1100])
    plt.plot(input_data[:,0],input_data[:,1],'ro')

    #initial temperature
    t = 100

    #current length
    min_l = total_length(sol_arr,n)

    i = 0
    best_arr = []
    while t > 0.1:
        i = i + 1

        #two_opt method- for optimization
        sol_arr = two_opt_optimization(sol_arr,t,n)

        #after 200 steps restart the process until the temperature is less than 0.1
        if i >= 200 :
            i = 0
            current_l = total_length(sol_arr,n)

            #keeping the cooling schedule slow
            t = t*0.9995

            if current_l < min_l:
                min_l = current_l
                best_arr = sol_arr[:]

    return best_arr

def main():
    start = time.time()
    n = 0
    # creating random inputs
    print("أدخل كمية المدخلات:")
    n = input()
    input_data = np.random.randint(1000,size =(int(n),2))
    final_arr = sa_algorithm(input_data)
    final_l = total_length(final_arr,int(n))
    end = time.time()
    
    print("\nأقصر طول باستخدام خوارزمية سميوليتد انيلينق:", final_l)
    print("\nمقياس الوقت: %.2f ثانية" % (end-start))
    print("\nتمثيل بياني للحل الأمثل:\n")

if __name__ == "__main__":
    main()          
```
